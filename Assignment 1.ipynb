# Web Scrapping

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://myanimelist.net/topmanga.php'
```

```python
res = requests.get(url)
```

```python
res
```

```python
siro = BeautifulSoup(res.text)
```

```python
siro
```

```python
Manga = siro.select('.fs14.fw-b')
```

```python
Manga
```

```python
Manga_title = []
```

```python
for item in Manga:
    Manga_title.append(item.text)
```

```python
Manga_title
```

```python
Volumn = siro.select('.mt4')
```

```python
Volumn
```

```python
release_volumn =[]
```

```python
for item in Volumn:
    release_volumn.append(item.text)
```

```python
release_volumn
```

```python
Score = siro.select('.on.score-label')
```

```python
Score
```

```python
release_score = []
```

```python
for item in Score:
    release_score.append(item.text)
```

```python
release_score
```

```python
import pandas as rc
```

```python
ff = rc.DataFrame(Manga_title,columns=['Manga'])
```

```python
ff
```

```python
ff['Volumn'] = Volumn
```

```python
ff
```

```python
ff['Score'] = Score
```

```python
ff
```

```python
ff.to_csv('ff.csv')
```
